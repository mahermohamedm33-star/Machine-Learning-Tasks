```python
import numpy as np
from sklearn.linear_model import LinearRegression, LogisticRegression

x_linear = np.array([[2],[4],[6],[8]])
y_linear = np.array([10,20,30,40])

linear_model = LinearRegression()
linear_model.fit(x_linear, y_linear)

test_number_linear = 40
linear_result = linear_model.predict([[test_number_linear]])

print(f'Linear Regression Result for {test_number_linear} = {linear_result[0]}')
```

```python
x_logistic = np.array([[10],[20],[30],[40],[60],[80]])
y_logistic = np.array([0,0,0,1,1,1])

logistic_model = LogisticRegression()
logistic_model.fit(x_logistic, y_logistic)

test_number_logistic = 34
logistic_result = logistic_model.predict([[test_number_logistic]])

print(f'Logistic Regression Result for {test_number_logistic} = {logistic_result[0]}')
```
